```python
#importing datetime library to convert date columns from string to datetime
import datetime
```

```python
data['NCH_BENE_DSCHRG_DT'] = pd.to_datetime(data['NCH_BENE_DSCHRG_DT'], format = '%Y%m%d')
data['CLM_ADMSN_DT'] = pd.to_datetime(data['CLM_ADMSN_DT'], format = '%Y%m%d')
```

```python
#subtracting those columns to determine time between start and end of claim
TimeOfStay = data['NCH_BENE_DSCHRG_DT'] - data['CLM_ADMSN_DT']
```

```python
TimeOfStay
```

```python
#convert to purely numeric value without "days" label
TimeOfStayFloat = TimeOfStay.dt.days
```

```python
TimeOfStayFloat
```

```python
#Remove those datetime columns from data
data.drop(['CLM_FROM_DT', 'CLM_THRU_DT', 'CLM_ADMSN_DT', 'NCH_BENE_DSCHRG_DT'], axis = 1, inplace = True)
```

```python
TimeOfStayFloat.median()
```

```python
#Add the new claims time period to data
data.insert(13, "Time of Stay", TimeOfStayFloat, True)
```
